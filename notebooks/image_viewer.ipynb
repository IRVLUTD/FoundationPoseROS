# Load Mask Image

```python
def colorize_mask(mask):
    mask = mask.astype(np.uint8)
    mask_color = np.zeros((mask.shape[0], mask.shape[1], 3), dtype=np.uint8)
    for i in range(1, 255):
        mask_color[mask == i] = np.random.randint(0, 255, size=(3,))
    return mask_color
img = cv2.imread("../recordings/segmentation/init_segmentation/mask_color_000000.png", cv2.IMREAD_GRAYSCALE)
print(np.unique(img))
plt.imshow(colorize_mask(img))
```
